```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
encoded_data=df.copy()
```

```python
label_encoder=LabelEncoder()
encoded_data['Gender_Label'] = label_encoder.fit_transform(encoded_data['Gender'])
```

```python
one_hot_encoded=pd.get_dummies(encoded_data['Department'],prefix='Department')
encoded_data=pd.concat([encoded_data,one_hot_encoded],axis=1)
```

```python
Education_mapping={'PhD': 'First','Masters': 'Second','Bachelors': 'Third'}
encoded_data['Education_category']=encoded_data['Education'].map(Education_mapping)
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
file_path='/content/preprocessing_practice_dataset.csv'
preprocessing_dataset=pd.read_csv(file_path)
```

```python
numerical_column = ['Age','Experience_Years','Performance_Rating']
```

```python
preprocessing_dataset.dropna(subset=numerical_column,inplace=True)
```

```python
minmax_scaler=MinMaxScaler()
standard_scaler=StandardScaler()
```

```python
preprocessing_dataset['Age_MinMax']=minmax_scaler.fit_transform(preprocessing_dataset[['Age']])
preprocessing_dataset['Experience_MinMax']=minmax_scaler.fit_transform(preprocessing_dataset[['Experience_Years']])
preprocessing_dataset['Age_Standard']=standard_scaler.fit_transform(preprocessing_dataset[['Age']])
preprocessing_dataset['Experience_Standard']=standard_scaler.fit_transform(preprocessing_dataset[['Experience_Years']])
```
