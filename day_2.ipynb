```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
endata=df.copy()
```

```python
label_encoder=LabelEncoder()
endata['age_Label']=label_encoder.fit_transform(endata['age'])
```

```python
one_hot_encoded=pd.get_dummies(endata['rad'],prefix='rad')
endata=pd.concat([endata,one_hot_encoded],axis=1)
```

```python
chas_mapping={0:'First',1: 'Second'}
endata['chas_Category']=endata['chas'].map(chas_mapping)
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
file_path='/content/boston.csv'
bst_data=pd.read_csv(file_path)
```

```python
numerical_columns=['zn','rad','tax']
```

```python
bst_data.dropna(subset=numerical_columns,inplace=True)
```

```python
minmax_scaler=MinMaxScaler()
standard_scaler=StandardScaler()
```

```python
bst_data['zn_MinMax']=minmax_scaler.fit_transform(bst_data[['zn']])
bst_data['rad_MinMax']=minmax_scaler.fit_transform(bst_data[['rad']])
bst_data['zn_Standard']=standard_scaler.fit_transform(bst_data[['zn']])
bst_data['rad_Standard']=standard_scaler.fit_transform(bst_data[['rad']])
```
